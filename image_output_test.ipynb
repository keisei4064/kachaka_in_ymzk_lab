```python
from IPython.display import display, Image
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image as PILImage
import os
import shutil

# 
output_dir = 'output_frames/'

# 
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir)

# 
def save_fig(index):
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x + index * 0.1)

    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_title(f'Frame {index}')
    fig.savefig(os.path.join(output_dir, f'figure_{index}.png'))
    plt.close()

# 
num_frames = 10
for i in range(num_frames):
    save_fig(i)

# 
image_list = [PILImage.open(os.path.join(output_dir, f'figure_{i}.png')) for i in range(num_frames)]

# GIF
image_list[0].save('output_animation.gif', save_all=True, append_images=image_list[1:], duration=200, loop=0)
with open('output_animation.gif', 'rb') as f:
    display(Image(data=f.read(), format='png'))

# GIF
# shutil.rmtree(output_dir)
# os.remove('output_animation.gif')
```
