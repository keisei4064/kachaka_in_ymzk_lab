```python
%matplotlib ipympl

import matplotlib.pyplot as plt
import numpy as np

# ボールの初期位置と半径
ball1_pos = np.array([1, 1])
ball1_radius = 0.5

ball2_pos = np.array([2, 2])
ball2_radius = 0.5

# ボール同士の距離を計算
distance = np.linalg.norm(ball1_pos - ball2_pos)

# 接触判定
if distance < ball1_radius + ball2_radius:
    print("ボールが接触しています。")
else:
    print("ボールは接触していません。")

# Matplotlibを使用してボールを描画
fig, ax = plt.subplots()
circle1 = plt.Circle(ball1_pos, ball1_radius, color="r", fill=False)
circle2 = plt.Circle(ball2_pos, ball2_radius, color="b", fill=False)

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect("equal", adjustable="box")  # アスペクト比を保持して描画

plt.xlim(0, 3)
plt.ylim(0, 3)
plt.grid(True)
plt.show()
```
