```python
from IPython.display import display, Image
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image as PILImage
import os
import shutil

# 出力ディレクトリのパス
output_dir = 'output_frames/'

# 出力ディレクトリが存在する場合、削除して再作成
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir)

# 図を保存する関数
def save_fig(index):
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x + index * 0.1)

    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_title(f'Frame {index}')
    fig.savefig(os.path.join(output_dir, f'figure_{index}.png'))
    plt.close()

# 図を複数回保存
num_frames = 10
for i in range(num_frames):
    save_fig(i)

# 保存した図を読み込んでアニメーションのための画像リストを作成
image_list = [PILImage.open(os.path.join(output_dir, f'figure_{i}.png')) for i in range(num_frames)]

# GIFを表示
image_list[0].save('output_animation.gif', save_all=True, append_images=image_list[1:], duration=200, loop=0)
with open('output_animation.gif', 'rb') as f:
    display(Image(data=f.read(), format='png'))

# アニメーション表示後に出力ディレクトリとGIFファイルを削除
# shutil.rmtree(output_dir)
# os.remove('output_animation.gif')
```

```python
from IPython.display import display, Image, clear_output
import matplotlib.pyplot as plt
import numpy as np

# アニメーションのフレーム数
num_frames = 10

for i in range(num_frames):
    # グラフの作成
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x + i * 0.2)

    plt.plot(x, y)
    plt.title(f'Frame {i + 1}')

    # グラフを画像として保存
    plt.savefig(f'animation_frame_{i + 1}.png')

    # 表示をクリア
    clear_output(wait=True)
    # 画像を表示
    display(Image(filename=f'animation_frame_{i + 1}.png'))


    # 描画をクリア
    plt.clf()
```
